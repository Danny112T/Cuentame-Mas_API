```python
from mlx_lm import load, generate
```

```python

MODEL = "../original_models/Meta-Llama-3.1-8B-Instruct-bf16/"
PROMPT = "¿Sabes alguna historia de Zacatecas, México?"
```

```python

model, tokenizer = load(MODEL)

messages = [{"role": "user", "content": PROMPT}]
```

```python
prompt = tokenizer.apply_chat_template(
    messages, tokenize=False, add_generation_prompt=True
)
```

```python

response = generate(
    model,
    tokenizer,
    prompt,
    verbose=True,
    max_tokens=512
    )

print(type(response))

```
